# Visualization Practice, Wine Properties

```python
import pandas as pd
my_white_wine_data = pd.read_csv("winequality-white.csv", sep = ";")
my_red_wine_data = pd.read_csv("winequality-red.csv", sep = ";")
```

```python
my_red_wine_data.describe()
```

### Pose a question
#### #What qualities are similar or dissimilar between red and white wine(s)

```python
# Build a correlation matrix for red and white wine
# 1.0 is a perfect correlation, if no relationship correlation is 0, 0.5 is good
# Negative correlation, as one value goes up the other goes down
my_white_wine_data.corr()
my_red_wine_data.corr()
```

### Visualization: correlation matrix side by side
##### create correlation matrices

```python
import seaborn as sns

red_wine = my_red_wine_data.corr()
white_wine = my_white_wine_data.corr()
```

```python
pip install seaborn
```

```python
#red wine

sns.heatmap(red_wine)
```

```python
#white wine

sns.heatmap(white_wine)
```

#### Plot individual variables for emphasis

```python
#use plotly
#use plotly
# my_white_wine_data is the df

import plotly.express as px

px.scatter(my_white_wine_data, x = "citric acid", y = "fixed acidity")
```

```python
#use plotly: scatter plot with trend line
# the trend line slope is the model for the information

import plotly.express as px

px.scatter(my_red_wine_data, x = "citric acid", y = "fixed acidity",
           
          color_discrete_sequence = ['green'],
          opacity = 0.2,
          labels={
                     "citric acid": "Citric acid",
                     "fixed acidity": "Fixed acidity"                    
                 },
          title = "Correlation between Citric acid and Fixed acidity in Red Wine",
          trendline="ols")
```
