```python
import pandas as pd
```

```python
#Load data
my_data = pd.read_excel('customer_data_set.xlsx')
```

```python
my_data.head(10)
```

```python
# Age range
# Find the minimum and maximum values in the column
min_age = my_data['Age'].min()
max_age = my_data['Age'].max()

# Calculate the range
age_range = max_age - min_age

print("Range for ages:", age_range)
```

```python
# Get rid of NAN
condition = my_data['Age'].isna()
# Apply the condition to filter out the rows with bad data
cleaned_df = my_data[~condition]
cleaned_df.head(10)
```

```python
# Sort the DataFrame by the specified column in descending order
sorted_df = cleaned_df.sort_values(by='Age', ascending=False)
sorted_df
sorted_df.head(10)
```

```python
# Delete the row in-place
sorted_df = sorted_df.drop('239')
sorted_df.head(10)
```
