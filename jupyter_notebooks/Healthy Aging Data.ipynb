```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
national_poll_on_healthy_aging_npha = fetch_ucirepo(id=936) 
  
# data (as pandas dataframes) 
X = national_poll_on_healthy_aging_npha.data.features 
y = national_poll_on_healthy_aging_npha.data.targets 
  
# metadata 
print(national_poll_on_healthy_aging_npha.metadata) 
  
# variable information 
print(national_poll_on_healthy_aging_npha.variables) 
```

```python
# Calculate statistics
mean = X.mean()
std_dev = X.std()
variance = X.var()
range_ = X.max(numeric_only=True) - X.min(numeric_only=True)
median = X.median()
mode = X.mode().loc[0]

print('Mean:\n', mean)
print('\nStandard Deviation:\n', std_dev)
print('\nVariance:\n', variance)
print('\nRange:\n', range_)
print('\nMedian:\n', median)
print('\nMode:\n', mode)
```

```python
#For a single column
# Select the column
column = X['Bathroom_Needs_Keeps_Patient_from_Sleeping']

# Calculate statistics
mean = column.mean()
std_dev = column.std()
variance = column.var()
range_ = column.max() - column.min()
median = column.median()
mode = column.mode()[0]

print('Mean:', mean)
print('Standard Deviation:', std_dev)
print('Variance:', variance)
print('Range:', range_)
print('Median:', median)
print('Mode:', mode)
```
